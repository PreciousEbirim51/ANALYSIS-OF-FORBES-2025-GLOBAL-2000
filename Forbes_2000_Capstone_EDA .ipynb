INDUSTRY LEVEL ANALYSIS

```python
num_cols = ['Revenue ($)', 'Profits ($)', 'Assets ($)', 'Market Value ($)']
```

```python
industry_summary = (
    df.groupby('Industry')[num_cols]
      .agg(['mean', 'median', 'sum'])
      .sort_values(('Revenue ($)', 'sum'), ascending=False)
)

industry_summary.head(10)
```

```python
industry_summary = (
    df.groupby('Industry')['Profits ($)']
      .agg(['mean', 'median', 'sum'])
      .sort_values(( 'sum'), ascending=False)
)

industry_summary.head(10)
```

```python
industry_summary = (
    df.groupby('Industry')['Revenue ($)']
      .agg(['mean', 'median', 'sum'])
      .sort_values(( 'sum'), ascending=False)
)

industry_summary.head(10)
```

```python
industry_summary = (
    df.groupby('Industry')['Assets ($)']
      .agg(['mean', 'median', 'sum'])
      .sort_values(( 'sum'), ascending=False)
)

industry_summary.head(10)
```

```python
industry_summary = (
    df.groupby('Industry')['Market Value ($)']
      .agg(['mean', 'median', 'sum'])
      .sort_values(( 'sum'), ascending=False)
)

industry_summary.head(10)
```

COUNTRY / HEADQUARTER ANALYSIS

```python
country_summary = (
    df.groupby('Headquarters')[num_cols]
      .sum()
      .sort_values('Revenue ($)', ascending=False)
)

country_summary.head(10)
```

```python
country_grouped = df.groupby('Headquarters')[num_cols].sum()

for col in num_cols:
    print(f"\nTop 10 Headquarters by {col}")
    display(
        country_grouped.sort_values(col, ascending=False).head(10)
    )
```
